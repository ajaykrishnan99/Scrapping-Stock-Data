```python
import urllib.request
url='https://www1.nseindia.com/content/historical/EQUITIES/2021/AUG/cm25AUG2021bhav.csv.zip'
urlpart1='https://www1.nseindia.com/content/historical/EQUITIES/'
year=['2019','2020','2021']
date=[str(x).zfill(2) for x in range(1,32)]
month=['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
filenames=[]
```

```python
# year[0]+month[0]+date[0]
```

```python
filenames
```

```python
for y in year:
    for m in month:
        for d in date:
            try:
                filename='cm'+d+m+y+'bhav.csv.zip'
                url_main=urlpart1+y+'/'+m+'/cm'+d+m+y+'bhav.csv.zip'
                print(url_main)
                urllib.request.urlretrieve(url_main,'/Users/Ajay/Ajay_RMIT/Project_try/STOCK_PREDICTION/StockDownload/'+str(filename))
                print(url_main)
                filenames.append(filename)
            except:
                url_main=urlpart1+y+'/'+m+'/cm'+d+m+y+'bhav.csv.zip'

                print("Exception:",url_main)
#                 print('Exception')
```

```python
import zipfile
import os
for i in filenames:
    path_to_zip_file='StockDownload/'+i
    print(path_to_zip_file)
    with zipfile.ZipFile(path_to_zip_file, 'r') as zip_ref:
        zip_ref.extractall('StockDownload/')
    os.remove(path_to_zip_file)
```

```python
filenames_wzip=[x.split('.')[0]+'.'+x.split('.')[1] for x in filenames]
```
